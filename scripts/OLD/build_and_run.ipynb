```python
city = "Jeddah"
for year in [ "2015","2016", "2017", "2018" ]:
    with open("F:\Non_Technical_loss_Analysis\Jeddah_Data_for_Non_Technical_loss_Analysis/Jeddah_Transaction_Data_" + year + ".txt") as f:
        counter = 0
        header, data = [], []
        currentMonth = None
        for line in f:
            line = line.replace("\n", "")
            if counter:
                month = line[4:6]
                if month != currentMonth:
                    if currentMonth is not None:
                        with open("C:\Users\Abdullah\Desktop/Data/" + city + "/" + year + "/ "+ currentMonth + ".csv", "w+") as f2:
                            f2.write(header[0])
                            for h in header[1:]:
                                f2.write( "," + h)
                            for row in data:
                                f2.write( "\n" + row )
                        data = []
                    currentMonth = month
                data.append( line.replace(";", ","))
            else:
                header = line.replace('"', "").split(";")
            counter += 1
    with open("C:\Users\Abdullah\Desktop/Data/" + city + "/" + year + "/" + currentMonth + ".csv", "w+") as f2:
            f2.write(header[0])
            for h in header[1:]:
                f2.write( "," + h)
            for row in data:
                f2.write( "\n" + row )
```
